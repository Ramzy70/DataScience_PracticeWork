## **1)**

```python
df = pd.read_csv("/kaggle/input/amazon-ratings/ratings_Beauty.csv")
```

## **2)**

```python
df_sorted = df.sort_values(by='Rating', ascending=False)
top_10_products = df_sorted.head(10)
print(top_10_products)
```

## **3)**

```python
df['Days'] = pd.to_datetime(df['Timestamp'], unit='s')
print(df.head())
```

## **4)**

```python
import matplotlib.pyplot as plt

# Calculate the average rating for each day
average_ratings = df.groupby('Days')['Rating'].mean()

# Plotting
plt.figure(figsize=(15, 6))
plt.plot(average_ratings.index, average_ratings.values, marker='o', linestyle='-')
plt.title('Average Rating Over Time')
plt.xlabel('Days')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

## **5)**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample 1% of the data
sampled_df = df.sample(frac=0.01)

# Group by 'Days' and 'ProductId' to get the count of products sold each day
product_counts = sampled_df.groupby(['Days', 'ProductId']).size().unstack(fill_value=0)

# Plotting
plt.figure(figsize=(10, 6))  # Adjust the figure size

# Define a color map for each product
colors = plt.cm.rainbow(range(len(product_counts.columns)))

for i, product_id in enumerate(product_counts.columns):
    plt.plot(product_counts.index, product_counts[product_id], label=f'Product {product_id}', marker='o', color=colors[i])

plt.title('Number of Products Sold by Day (1% Sample)')
plt.xlabel('Days')
plt.ylabel('Number of Products Sold')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.grid(True)

# Save the plot as an image file (e.g., PNG)
plt.savefig('product_sales_plot.png')

# Display the file path to the saved image
print("Plot saved as product_sales_plot.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with 'Rating' and other relevant columns

# Count the occurrences of each rating
rating_counts = df['Rating'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(rating_counts.index, rating_counts.values, color='skyblue')

plt.title('Number of Occurrences of Each Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rating_counts.index)
plt.grid(axis='y')
plt.show()
```
